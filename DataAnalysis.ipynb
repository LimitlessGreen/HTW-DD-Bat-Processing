```python
from bokeh.models import ColumnDataSource, FactorRange, DateRangeSlider
from bokeh.plotting import figure, show
from bokeh.sampledata.commits import data
from bokeh.transform import jitter
from bokeh.io import output_notebook
from datetime import datetime, time


data = my_proj.get_data()
data = data.fillna(0)
data = data[(data[['Aufnahme']] != 0).all(axis=1)]
data = data [ (data['Manual_ID'] != "unbestimmt") & (data['Manual_ID'] != "keine_Art") ]


y = [*set(data["Manual_ID"].values)]


source = ColumnDataSource(data)

output_notebook()

p = figure(width=1000, height=500, y_range=y, x_axis_type='datetime', 
           title="Abhngigkeit des Artauftretens von der Uhrzeit")

min_time = data["Time"]

filtered_data_start = data.loc[(data["Time"] >= datetime.strptime('12:00', '%H:%M').time())
                         & (data["Time"] <= datetime.strptime('23:00', '%H:%M').time())]

filtered_data_end = data.loc[(data["Time"] >= datetime.strptime('01:00', '%H:%M').time())
                         & (data["Time"] <= datetime.strptime('12:00', '%H:%M').time())]

start_time = filtered_data_start["Time"].min()
end_time = filtered_data_end["Time"].max()

slider = DateRangeSlider(value=(start_time, end_time))

p.scatter(x='Time', y=jitter('Manual_ID', width=.7, range=p.y_range),  source=source, alpha=0.15)
# p.scatter(x='DateTime', y="Sozialruf",  source=source, alpha=0.6, size=20)

p.xaxis.formatter.days = ['%Hh']
p.x_range.range_padding = 0
p.ygrid.grid_line_color = None

#p.x_range.start = start_time
#p.x_range.end = end_time
show(p)
```
